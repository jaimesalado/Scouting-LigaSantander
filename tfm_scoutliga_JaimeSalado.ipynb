## 3. Descripci√≥n de los datos.

En nuestro conjunto de datos utilizado nos encontraremos con diferentes campos de informaci√≥n que describiremos a continuaci√≥n : 

* **Id** : Contiene el identificador de nuestra fila de datos.
* **Nombre** : Contiene nombre de jugador del cual mostraremos informaci√≥n.
* **Posici√≥n** : Contiene la zona donde juega el jugador : Portero, Defensa, Mediocampista, Delantero.
* **Apariciones** : Contiene la cantidad de partidos en los  que el jugador particip√≥ durante la temporada. 
* **Minutos** : Contiene la cantidad total de minutos jugados durante la temporada.
* **Goles** : Contiene la cantidad total del goles marcados durante la temporada por el jugador.
* **Tiros totales** : Contiene el n√∫mero de tiros realizados durante la temporada por el jugador.
* **Conversi√≥n goles %** : Contiene porcentaje de efectividad de cara a gol por parte del jugador durante la temporada.
* **Regates Exito %** : Contiene porcentaje de efectividad de los regates realizados por el jugador durante la temporada.
* **Perdida Posesi√≥n** : Contiene el n√∫mero de veces que el jugador perdi√≥ el bal√≥n durante la temporada.
* **Asistencias** : Contiene el n√∫mero de pases decisivos para la realizaci√≥n de un gol por parte del jugador durante la temporada.
* **Pases Total** : Contiene la cantidad total de pases realizados por el jugador durante la temporada.
* **Pases Claves** : Contiene n√∫mero total de √∫ltimos pases que conduce a un tiro a porter√≠a de un compa√±ero durante la temporada.
* **Pases Precisos %**: Contiene porcentaje de efectividad de pases realizados por el jugador durante la temporada.
* **Entradas** : Contiene el n√∫mero total de entradas realizadas por el jugador para quitar la posesi√≥n al rival durante la temporada.
* **Intercepciones** : Contiene cantidad total de veces que el jugador durante la temporada corta,desv√≠a o bloquea un pase del rival.
* **Total Duelos Ganados** : Contiene cantidad total de veces durante la temporada que un jugador gana el bal√≥n en la disputa ante un rival.
* **Total Duelos Ganado %** : Contiene porcentaje de efectividad de disputas ganadas por el jugador durante la temporada.
* **Faltas** : Contiene n√∫mero total de infracciones cometidas por el jugador durante la temporada.  
* **Tarjeta Amarilla** : Contiene n√∫mero total de amonestaciones por parte del jugador durante la temporada.
* **Tarjeta Roja** : Contiene n√∫mero total de expulsiones por parte del jugador durante la temporada.
* **Paradas Partido %** : Contiene porcentaje de efectividad de paradas realizadas por partido de un jugador durante la temporada.(portero) 	
* **Goles Encajados** : Contiene la cantidad total de goles recibidas por el jugador durante la temporada. (portero)	
* **Imbatido** : Contiene n√∫mero de partidos donde el jugador no recibi√≥ gol durante la temporada.(portero)	
* **Rendimiento** : Contiene nota final del jugador dependiendo del desempe√±o realizado durante la temporada.

## 8. Entrenamiento del modelo y comprobaci√≥n del rendimiento.

Tras el estudio,visualizaci√≥n y preparaci√≥n de nuestros datos ahora se dividir√°n los datos en conjuntos de entrenamiento y prueba.
Luego utilizaremos nuestros datos de entrenamiento y prueba para realizar modelos de predicci√≥n con algoritmos de aprendizaje autom√°tico seleccionados y se evaluar√° el rendimiento que han tenido cada uno de ellos seleccionando el m√°s adecuado

### 8.1. Elegir campos del modelo :

Elegimos los campos que utilizaremos en los diferentes modelos y lo guardaremos en una variable llamada datosmodelo.

* (Posici√≥n,minutos,goles,conversi√≥n de goles %, regates √©xito%, perdida posesi√≥n, asistencias, pases claves, pases precisos %, entradas, intercepciones, total duelos ganados %, faltas, paradas partido %, goles encajados, rendimiento)


Tambi√©n dividimos el conjunto de datos entrenamiento y prueba, donde utilizamos el 20 % de datos como prueba y para evaluar el modelo  y el 80 % restante se utiliza como entrenamiento.

Ademas establecemos una semilla con random_state= 42 es un par√°metro utilizado para la divisi√≥n de los datos, cada vez que se ejecute el c√≥digo se obtiene la misma divisi√≥n de los datos y es √∫til para asegurarse de que los resultados obtienen los mismos resultados cada vez que se ejecuta el c√≥digo utilizando mismos datos de entrada y par√°metros y tambi√©n obtenga resultados similares)

```python
from sklearn.model_selection import train_test_split

datosmodelo = laliga[['Posicion', 'Minutos', 'Goles', 'Conversion goles %', 'Regates Exito %', 'Perdida Posesion', 
         'Asistencias', 'Pases Claves', 'Pases Precisos %', 'Entradas', 'Intercepciones', 
         'Total Duelos Ganado %', 'Faltas', 'Paradas Partido %', 'Goles Encajados', 'Rendimiento']]

# Divide los datos en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(datosmodelo.drop('Rendimiento', axis=1), datosmodelo['Rendimiento'], 
                                                    test_size=0.2, random_state=42)
```

### 8.2. Entrenar modelo de regresi√≥n lineal :

El primer modelo que vamos a probar es el de regresi√≥n lineal, es utilizado en el aprendizaje autom√°tico y estad√≠sticas para predecir una variable num√©rica continua a partir de una o varias variables independientes, su objetivo es encontrar una linea recta que ajuste los datos para poder hacer predicciones precisas.

A continuaci√≥n implementamos el modelo de regresi√≥n lineal utilizando la biblioteca   Scikit-learn (sklearn) para entrenar y ajustar nuestro modelo, lo introducimos en una variable llamada reg y llamamos a esa variable para entrenar con fit.

Una vez ajustado el modelo ha sido ajustado se puede utilizar para hacer predicciones sobre nuevos datos. 



```python
from sklearn.linear_model import LinearRegression

# Entrena un modelo de regresi√≥n lineal
reg = LinearRegression()
reg.fit(X_train, y_train)
```

#### 8.2.1 Valor de predicciones.

A continuaci√≥n vamos a hacer unas predicciones


```python
prediccioneslineal = reg.predict(X_test)
prediccioneslineal
```

#### 8.2.2. Porcentaje de error y coeficiente de determinaci√≥n:

Ahora evaluamos la calidad del modelo de regresi√≥n lineal y se determina su capacidad para predecir valores de rendimiento.

El porcentaje de error y el coeficiente de determinaci√≥n son dos medidas de evaluaci√≥n comunes utilizadas en la regresi√≥n lineal para evaluar el rendimiento del modelo. 
El porcentaje de error indica qu√© tan cerca est√° una predicci√≥n de su valor real.Cuanto menor sea el porcentaje de error, m√°s precisa ser√° la predicci√≥n

El coeficiente de determinaci√≥n mide como se ajustan los datos a la l√≠nea de regresi√≥n.Su valor var√≠a entre 0 y 1.Cercano a 1 indica que el modelo ajusta muy bien los datos, mientras que un valor cercano a 0 indica que el modelo no ajusta bien los datos.



```python
from sklearn.metrics import mean_squared_error, r2_score

error = np.sqrt(mean_squared_error(y_test, prediccioneslineal))
error_porcentual = error / np.mean(y_test) * 100

print("Error porcentual : %f" % error_porcentual)

# coeficiente de determinacion 1 es la prediccion perfecta
print("coeficiente de determinacion: ", r2_score(y_test, prediccioneslineal))
```
